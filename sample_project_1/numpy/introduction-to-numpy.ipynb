## Dot product

```python
np.random.seed(0)

mat1 = np.random.randint(10, size=(5,3))
mat2 = np.random.randint(10, size=(5,3))
```

```python
mat1
```

```python

mat2
```

```python
mat1.shape, mat2.shape
```

```python
#This is element wise multiplication (Hadamard product)
mat1 * mat2
```

Dot product is like row by column so the length of the row of the first be equal to the length of the column of the second. In simple terms,the numbers on inside which are closer to the operator must be the same.  e.g (3x3 * 3x2) 

```python
#np.dot(mat1, mat2)
```

To make this work, we could use a transpose to invert the dimensiosn and make the inner numbers similar

```python
mat1.shape, mat2.T.shape
```

```python
mat_3 = np.dot(mat1, mat2.T)
mat_3
```

```python
#the shape will be the combo of the outer numbers
mat_3.shape
```

## Dot product example (nut butter sales)

```python
np.random.seed(0)
#Number of jars sold
sales_amounts = np.random.randint(20, size=(5,3))
sales_amounts
```

```python
# create weekly_sales DataFrame
weekly_sales = pd.DataFrame(sales_amounts,
                           index = ["Mon", "Tues", "Wed", "Thurs", "Fri"],
                           columns=["Almond Butter", "Peanut Butter", "Cashew Butter"])
weekly_sales
```

```python
#Create prices array
prices = np.array([10, 8, 12])
prices
```

```python
# Create butter prices dataframe
butter_prices = pd.DataFrame(prices.reshape(1, 3),
                            index = ["Price"],
                            columns=(["Almond Butter", "Peanut Butter", "Cashew Butter"]))
butter_prices
```

```python
total = np.dot(butter_prices, weekly_sales.T)
total
```

```python
# Create daily sales
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
#gotta traspose daily_sales since its a row but we want a column
weekly_sales["Total ($)"] = daily_sales.T
weekly_sales
```
